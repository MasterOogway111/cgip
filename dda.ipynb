```python
import matplotlib.pyplot as plt

def dda_without_slope_cases(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1

    steps = max(abs(dx), abs(dy))

    x_increment = dx / steps
    y_increment = dy / steps

    x = x1
    y = y1

    x_points = []
    y_points = []

    for i in range(steps + 1):
        x_points.append(round(x))
        y_points.append(round(y))
        x += x_increment
        y += y_increment

    return x_points, y_points


x1 = int(input("Enter x1: "))
y1 = int(input("Enter y1: "))
x2 = int(input("Enter x2: "))
y2 = int(input("Enter y2: "))

x_pts, y_pts = dda_without_slope_cases(x1, y1, x2, y2)

plt.scatter(x_pts, y_pts)          # plot points
plt.plot(x_pts, y_pts)             # join the dots
plt.title("DDA Line Drawing (Without Slope Cases)")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

def dda_with_slope_cases(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1

    x_points = []
    y_points = []

    if dx == 0:
        y_step = 1 if y2 > y1 else -1
        for y in range(y1, y2 + y_step, y_step):
            x_points.append(x1)
            y_points.append(y)

    elif dy == 0:
        x_step = 1 if x2 > x1 else -1
        for x in range(x1, x2 + x_step, x_step):
            x_points.append(x)
            y_points.append(y1)

    else:
        slope = dy / dx

        if abs(slope) <= 1:
            steps = abs(dx)
            x_increment = 1
            y_increment = slope * x_increment

            x = x1
            y = y1

            for i in range(steps + 1):
                x_points.append(round(x))
                y_points.append(round(y))
                x += x_increment
                y += y_increment

        else:
            steps = abs(dy)
            y_increment = 1
            x_increment = (1 / slope) * y_increment

            x = x1
            y = y1

            for i in range(steps + 1):
                x_points.append(round(x))
                y_points.append(round(y))
                x += x_increment
                y += y_increment

    return x_points, y_points


x1 = int(input("Enter x1: "))
y1 = int(input("Enter y1: "))
x2 = int(input("Enter x2: "))
y2 = int(input("Enter y2: "))

x_pts, y_pts = dda_with_slope_cases(x1, y1, x2, y2)

plt.scatter(x_pts, y_pts)
plt.plot(x_pts, y_pts)
plt.title("DDA Line Drawing (With Slope Cases)")
plt.grid(True)
plt.show()
```
