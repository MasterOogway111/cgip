```python
import numpy as np
import matplotlib.pyplot as plt
import math


# 1. Input Control Points
control_points = np.array([
    [100, 200],
    [150, 300],
    [250, 250],
    [300, 350],
    [400, 200]
])

x_ctrl = control_points[:, 0]
y_ctrl = control_points[:, 1]


# 2. Apply Bezier Method
def compute_bezier(points, num_points=100):
    n = len(points) - 1

    t_vals = np.linspace(0, 1, num_points)
    curve = np.zeros((num_points, 2))

    for i, t in enumerate(t_vals):
        for j in range(n + 1):
            # Calculate Bernstein polynomial weight
            bernstein = (
                math.comb(n, j)
                * (t ** j)
                * ((1 - t) ** (n - j))
            )

            curve[i] += bernstein * points[j]

    return curve


bezier_curve = compute_bezier(control_points)


# 3. Visualization
plt.figure(figsize=(8, 5))

plt.plot(
    x_ctrl,
    y_ctrl,
    'o--',
    color='gray',
    label='Control Polygon'
)

plt.plot(
    bezier_curve[:, 0],
    bezier_curve[:, 1],
    color='green',
    linewidth=2,
    label='Bezier Curve'
)

plt.title("Task 2: Bezier Curve")
plt.legend()
plt.grid(True)

# Match screen coordinates
plt.gca().invert_yaxis()

plt.show()
```
