```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt


polygon = np.array([
    [1, 1],
    [4, 1],
    [4, 4],
    [1, 4]
])


def plot_polygon(original, transformed, title):
    plt.figure()

    orig = np.vstack([original, original[0]])
    trans = np.vstack([transformed, transformed[0]])

    plt.plot(orig[:, 0], orig[:, 1], marker='o', label="Original")
    plt.plot(trans[:, 0], trans[:, 1], marker='o', label="Transformed")

    plt.title(title)
    plt.legend()
    plt.grid()
    plt.axis('equal')

    plt.show()


plot_polygon(polygon, polygon, "Test Plot")
```

translation

```python
tx, ty = 3, 2
translated = polygon + [tx, ty]
plot_polygon(polygon, translated, "Translation")
```

```python
Sx, Sy = 2, 1.5
scaled = polygon * [Sx, Sy]

plot_polygon(polygon, scaled, "Scaling about Origin")
```

```python
theta = np.radians(45)
rotation_matrix = np.array([
[np.cos(theta), -np.sin(theta)],
[np.sin(theta), np.cos(theta)]
])
rotated = polygon @ rotation_matrix.T
plot_polygon(polygon, rotated, "Rotation about Origin")
```

```python
h, k = 4, 4
temp = polygon - [h, k]
rotated_temp = temp @ rotation_matrix.T
rotated_arbitrary = rotated_temp + [h, k]
plot_polygon(polygon, rotated_arbitrary, "Rotation about (4,4)")
```

```python
temp = polygon - [h, k]
scaled_temp = temp * [Sx, Sy]
scaled_arbitrary = scaled_temp + [h, k]
plot_polygon(polygon, scaled_arbitrary, "Scaling about (4,4)")
```

```python
comp = polygon * [Sx, Sy]
comp = comp @ rotation_matrix.T
comp = comp + [tx, ty]
plot_polygon(polygon, comp, "Composite Transformation")
```
