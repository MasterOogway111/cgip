```python
import numpy as np
import matplotlib.pyplot as plt


WIDTH = int(input("Enter screen width: "))
HEIGHT = int(input("Enter screen height: "))


# Buffers
z_buffer = np.full((HEIGHT, WIDTH), np.inf)
frame_buffer = np.zeros((HEIGHT, WIDTH, 3))


def draw_pixel(x, y, z, color):
    if 0 <= x < WIDTH and 0 <= y < HEIGHT:
        if z < z_buffer[y][x]:
            z_buffer[y][x] = z
            frame_buffer[y][x] = color


def draw_rectangle(x1, y1, x2, y2, z, color):
    for x in range(x1, x2):
        for y in range(y1, y2):
            draw_pixel(x, y, z, color)


# Input
n = int(input("Enter number of rectangles: "))

for i in range(n):
    print(f"\nRectangle {i + 1}")

    x1 = int(input("x1: "))
    y1 = int(input("y1: "))
    x2 = int(input("x2: "))
    y2 = int(input("y2: "))

    z = float(input("Depth (z): "))

    r = float(input("R (0-1): "))
    g = float(input("G (0-1): "))
    b = float(input("B (0-1): "))

    draw_rectangle(x1, y1, x2, y2, z, [r, g, b])


# Display
plt.imshow(frame_buffer)
plt.title("Z-Buffer Output")
plt.axis('off')
plt.show()
```
