```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow


def run_test_case():
    img_path = input("\nEnter image filename: ")

    img = cv2.imread(img_path)

    if img is None:
        print("Error: Image not found!")
        return

    print("\nEnter Window Coordinates:")
    xw_min = int(input("xw_min: "))
    yw_min = int(input("yw_min: "))
    xw_max = int(input("xw_max: "))
    yw_max = int(input("yw_max: "))

    print("\nEnter Viewport Coordinates:")
    xv_min = int(input("xv_min: "))
    yv_min = int(input("yv_min: "))
    xv_max = int(input("xv_max: "))
    yv_max = int(input("yv_max: "))

    xw_min, xw_max = sorted([xw_min, xw_max])
    yw_min, yw_max = sorted([yw_min, yw_max])

    xv_min, xv_max = sorted([xv_min, xv_max])
    yv_min, yv_max = sorted([yv_min, yv_max])

    window = img[yw_min:yw_max, xw_min:xw_max]

    Sx = (xv_max - xv_min) / (xw_max - xw_min)
    Sy = (yv_max - yv_min) / (yw_max - yw_min)

    resized = cv2.resize(
        window,
        (xv_max - xv_min, yv_max - yv_min)
    )

    canvas = np.zeros_like(img)
    canvas[yv_min:yv_max, xv_min:xv_max] = resized

    if Sx == 1 and Sy == 1:
        t = "No Scaling (Only Translation)"
    elif Sx > 1 or Sy > 1:
        t = "Upscaling"
    elif Sx < 1 or Sy < 1:
        t = "Downscaling"

    if Sx != Sy:
        t += " + Non-uniform Scaling (Stretching)"

    print("\nResults")
    print("Sx:", Sx)
    print("Sy:", Sy)
    print("Transformation:", t)

    cv2_imshow(canvas)


t = int(input("Enter number of test cases: "))

for i in range(t):
    print(f"\n--- Test Case {i + 1} ---")
    run_test_case()
```
