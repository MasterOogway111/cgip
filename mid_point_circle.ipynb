```python
def midpoint_circle(xc, yc, r):
    x = 0
    y = r
    p = 1 - r

    points = []

    while x <= y:
        # 8-way symmetry
        points.extend([
            (xc + x, yc + y),
            (xc - x, yc + y),
            (xc + x, yc - y),
            (xc - x, yc - y),
            (xc + y, yc + x),
            (xc - y, yc + x),
            (xc + y, yc - x),
            (xc - y, yc - x)
        ])

        x += 1

        if p < 0:
            p += 2 * x + 1
        else:
            y -= 1
            p += 2 * (x - y) + 1

    return points


# Input
xc = int(input("Center X: "))
yc = int(input("Center Y: "))
r = int(input("Radius: "))

points = midpoint_circle(xc, yc, r)

print("Circle Points:")
for point in points:
    print(point)
```
