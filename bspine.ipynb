import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import BSpline


# 1. Input Control Points
control_points = np.array([
    [100, 200],
    [150, 300],
    [250, 250],
    [300, 350],
    [400, 200]
])

x_ctrl = control_points[:, 0]
y_ctrl = control_points[:, 1]


# 2. Apply B-Spline Method
degree = 3  # Cubic B-Spline

num_points = len(control_points)

# Clamped knot vector ensures the curve touches
# the start and end points
knots = np.concatenate((
    [0] * (degree + 1),
    np.linspace(0.1, 0.9, num_points - degree - 1),
    [1] * (degree + 1)
))

# Generate the spline functions for X and Y
spline_x = BSpline(knots, x_ctrl, degree)
spline_y = BSpline(knots, y_ctrl, degree)

# Evaluate the curve
t_vals = np.linspace(0, 1, 100)

x_curve = spline_x(t_vals)
y_curve = spline_y(t_vals)


# 3. Visualization
plt.figure(figsize=(8, 5))

plt.plot(
    x_ctrl,
    y_ctrl,
    'o--',
    color='gray',
    label='Control Polygon'
)

plt.plot(
    x_curve,
    y_curve,
    color='red',
    linewidth=2,
    label='B-Spline Approximation'
)

plt.title("Task 3: B-Spline Approximation")
plt.legend()
plt.grid(True)

# Match screen coordinates
plt.gca().invert_yaxis()

plt.show()
