```python
import matplotlib.pyplot as plt
def plot_circle_points(xc, yc, x, y, points):
points.extend([
(xc + x, yc + y),
(xc - x, yc + y),
(xc + x, yc - y),
(xc - x, yc - y),
(xc + y, yc + x),
(xc - y, yc + x),
(xc + y, yc - x),
(xc - y, yc - x)
])
def bresenham_circle(xc, yc, r):
x = 0
y = r
d = 3 - 2*r
points = []
plot_circle_points(xc, yc, x, y, points)
while x <= y:
x += 1
if d < 0:
d = d + 4*x + 6
else:
y -= 1
d = d + 4*(x - y) + 10
plot_circle_points(xc, yc, x, y, points)
return points

xc = int(input("Enter xc: "))
yc = int(input("Enter yc: "))
r = int(input("Enter r: "))
points = bresenham_circle(xc, yc, r)
x_vals = [p[0] for p in points]
y_vals = [p[1] for p in points]

Enter xc: 0
Enter yc: 9
Enter r: 100
plt.scatter(x_vals, y_vals, s=5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Bresenham's Circle Drawing Algorithm")
plt.grid()
plt.show()
```
