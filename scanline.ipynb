```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon


class Edge:
    def __init__(self, ymax, x, invSlope):
        self.ymax = ymax
        self.x = x
        self.invSlope = invSlope


def scanline_fill(vertices):
    n = len(vertices)
    vertices = [(int(x), int(y)) for x, y in vertices]

    ymin_global = min(y for x, y in vertices)
    ymax_global = max(y for x, y in vertices)

    ET = [[] for _ in range(ymax_global + 1)]

    for i in range(n):
        x1, y1 = vertices[i]
        x2, y2 = vertices[(i + 1) % n]

        if y1 == y2:
            continue

        ymin = min(y1, y2)
        ymax = max(y1, y2)

        x = x1 if y1 < y2 else x2
        invSlope = (x2 - x1) / (y2 - y1)

        ET[ymin].append(Edge(ymax, x, invSlope))

    AET = []
    segments = []
    fill_points = []

    for y in range(ymin_global, ymax_global + 1):
        AET.extend(ET[y])

        AET = [e for e in AET if e.ymax != y]
        AET.sort(key=lambda e: e.x)

        for i in range(0, len(AET), 2):
            if i + 1 < len(AET):
                x1 = int(round(AET[i].x))
                x2 = int(round(AET[i + 1].x))

                segments.append((x1, x2, y))

                for x in range(x1, x2 + 1):
                    fill_points.append((x, y))

        for e in AET:
            e.x += e.invSlope

    return segments, fill_points, vertices


# Input vertices
n = int(input("Enter number of vertices: "))

vertices = []
print("Enter vertices (x y):")

for _ in range(n):
    x, y = map(int, input().split())
    vertices.append((x, y))


# Scanline filling
segments, fill_points, vertices = scanline_fill(vertices)


# Plotting
fig, ax = plt.subplots(figsize=(7, 7))

for x1, x2, y in segments:
    ax.plot([x1, x2], [y, y], linewidth=2)

if fill_points:
    fx, fy = zip(*fill_points)
    ax.scatter(fx, fy, s=20)

for (x, y) in fill_points:
    if len(fill_points) <= 30:
        ax.text(x + 0.1, y + 0.1, f"({x},{y})", fontsize=8)

polygon = Polygon(
    vertices,
    fill=False,
    edgecolor='black',
    linewidth=2
)

ax.add_patch(polygon)

ax.set_title("Scan Line Fill with Explicit Filled Points", fontsize=14)
ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.3)

plt.gca().invert_yaxis()
plt.show()
```
