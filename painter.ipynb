```python
import matplotlib.pyplot as plt


n = int(input("Enter number of polygons: "))

polygons = []

for i in range(n):
    print(f"\nPolygon {i + 1}")

    x1 = int(input("x1: "))
    y1 = int(input("y1: "))

    x2 = int(input("x2: "))
    y2 = int(input("y2: "))

    z = float(input("Depth (z): "))

    color = input("Color (e.g., red, blue): ")

    coords = [
        (x1, y1),
        (x2, y1),
        (x2, y2),
        (x1, y2)
    ]

    polygons.append((z, color, coords))


# Sort by depth (far → near)
polygons.sort(reverse=True)

fig, ax = plt.subplots()

for z, color, coords in polygons:
    x = [p[0] for p in coords] + [coords[0][0]]
    y = [p[1] for p in coords] + [coords[0][1]]

    ax.fill(x, y, color=color)

plt.title("Painter's Algorithm Output")
plt.gca().invert_yaxis()
plt.show()
```
