```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange


# 1. Input Control Points
control_points = np.array([
    [100, 200],
    [150, 300],
    [250, 250],
    [300, 350],
    [400, 200]
])

x_ctrl = control_points[:, 0]
y_ctrl = control_points[:, 1]


# 2. Apply Lagrange Interpolation
# Lagrange creates a polynomial that intersects every point exactly.
poly = lagrange(x_ctrl, y_ctrl)

# Generate smooth points for plotting
x_curve = np.linspace(min(x_ctrl), max(x_ctrl), 100)
y_curve = poly(x_curve)


# 3. Visualization
plt.figure(figsize=(8, 5))

plt.plot(
    x_ctrl,
    y_ctrl,
    'o--',
    color='gray',
    label='Control Polygon'
)

plt.plot(
    x_curve,
    y_curve,
    color='blue',
    linewidth=2,
    label='Lagrange Interpolation'
)

plt.title("Task 1: Lagrange Interpolation")
plt.legend()
plt.grid(True)

# Match screen coordinates
plt.gca().invert_yaxis()

plt.show()
```
